# Opioid Overdose & Income

Below is a scatterplot that visually shows the relationship between overdose rate and income levels of counties within the state of Virginia.

```python
""" Format of the scatterplot method is as follows: ax.scatter(x-series, y-series) """
fig, ax = plt.subplots()
""" 
ax.scatter() statement creates scatterplot
The 'alpha' parameter controls dot transparency: 1 = solid, <1 = various transparency levels, 0 = no mark
The c parameter designates color of the dots: 'b' stands for blue  
"""
ax.scatter(df_opi['MedianHouseholdIncome']/1000, df_opi['FPOO-Rate'], alpha=0.5, c = 'b')    # the 'alpha' parameter controls dot opacity
fig.suptitle('Opioid Overdoses vs. Income', fontsize = 16)
ax.xaxis.set_label_text('County Yearly Median Income (thousands)')
ax.yaxis.set_label_text('Opioid Overdose Rate')
fig.set_size_inches(11,8)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

## Question:
Do opioid overdoes tend to be associated with less affluent areasâ€”that is, areas where families have lower incomes?

### Opioid overdoses & Income

While many middle and lower income areas have few or no overdoses, the overdose rates for opioids are highest in lower income areas. Overdose rates appear to weakly cluster in middle income areas. While higher income areas tend to have fewer overdoses, they appear more likely to have at least some overdoses.

# Heroin Overdoses vs income

```python
""" Format of the scatterplot method is as follows: ax.scatter(x-series, y-series) """
fig, ax = plt.subplots()
""" 
ax.scatter() statement creates scatterplot
The 'alpha' parameter controls dot transparency: 1 = solid, <1 = various transparency levels, 0 = no mark
The c parameter designates color of the dots: 'b' stands for blue  
"""
ax.scatter(df_opi['MedianHouseholdIncome']/1000, df_opi['FPOO-Rate'], alpha=0.25, c = 'grey')
ax.scatter(df_opi['MedianHouseholdIncome']/1000, df_opi['FFHO-Rate'], alpha=0.5, c = 'b')    # the 'alpha' parameter controls dot opacity
fig.suptitle('Heroin Overdoses vs. Median Income',fontsize = 16)
ax.xaxis.set_label_text('County Yearly Median Income (thousands)')
ax.yaxis.set_label_text('Heroin Overdose Rate')
fig.set_size_inches(11,8)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

## Question:
Some people who start with opioid addictions are reported to transition to heroin use. What is the relationship in Virginia counties between opioid overdoses and heroin overdoses?

### Heroin overdoses & Income

The relationship between overdose rates and area income is less clear with heroin and fentanyl compared to opioid overdoses. Due to unpredictable purity, dosing is complicated and dangerous for heroin/fentanyl. While the highest rates are still located in lower and middle income areas, higher income areas have elevated rates of heroin/fentanyl overdose compared to opioid overdoses.

With the grey dots representing opioid overdoses, it is easy to find the same county's corresponding blue dot for heroin/fentanyl overdoses. This could either mean that addicts in those areas are moving from opioid use to heroin/fentanyl, or that they are using either depending on availability. What is clear is that in middle and higher income areas heroin/fentanyl overdose rates are higher than opioid overdose rates. This is likely because the cost burden is less limiting for addicts with higher income. 

## Question:

Which states are associated with the greatest number of United States presidents in terms of the presidentsâ€™ birthplaces?

```python
uniqueVals = list(df_presidents['State'].unique())
state_names = [i for i in uniqueVals]
data = []
for state in state_names:
    data.append(sum(df_presidents['State'] == state))
sort_me = list(zip(state_names,data))  
sort_me = sorted(sort_me, key=lambda x: x[1], reverse=True)
states,pres_count =zip(*sort_me)
fig,ax = plt.subplots()
fig.set_size_inches(16,8)
ax.bar(states,pres_count, width=0.75)
fig.suptitle('Presidents by Birth State',fontsize = 16)
ax.set_xlabel("Birth State")
plt.xticks(rotation=20)
ax.set_ylabel("Presidents")
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

### Analysis:

Virginia, Ohio, New York, and Massachusetts have most often been presidential birth states. Virginia, New York, and Massachusetts were among the most powerful colonies before the United States was founded. Ohio was aggressively settled and developed not long after the founding of the nation. In fact, at the onset of the American Revolution,  settlement of the Ohio territory was one of the strongest points of contention between powerful continental interests and those of the British empire.

It makes sense that more presidents have been born in some of the oldest and most powerful of the original states.

## Question:
How have vehicle sales in the United States varied over time?

```python
import matplotlib.dates as mdates
fig, ax = plt.subplots()
fig.suptitle('United States Audo Sales', fontsize = 16)
fig.set_size_inches(12,9)
ax.plot(df_NSA['DATE'][0:12],df_NSA['TOTALNSA'][0:12],color='orange')
plt.xticks(rotation=20)

plt.show()
```

```python
import statistics
months = list(df_NSA['DATE'][0:504])
month_sales = list(df_NSA['TOTALNSA'][0:504])
g_months = [months[x:x+12] for x in range(0, len(months),12)]
g_sales = [month_sales[x:x+12] for x in range(0, len(month_sales),12)]
years = []
for i in range(len(g_months)):
    years.append(1976+i)
sales = []
for i in range(len(g_sales)):
    sales.append(statistics.mean(g_sales[i]))
fig, ax = plt.subplots()
fig.suptitle('Average Monthly Auto Sales in US: 1976-2017', fontsize = 16)
fig.set_size_inches(11,7)
ax.plot(years,sales,color='orange')
ax.set_xlabel("Year", fontsize = 12)
ax.set_ylabel("Average Monthly Auto Sales", fontsize = 12)
plt.xticks(years,rotation=90)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_ylim([0, 1600])
plt.show()
```

## Analysis 

The year monthly average of auto sales has increased over time, interrupted by sharp dips that correspond with economic recessions.
